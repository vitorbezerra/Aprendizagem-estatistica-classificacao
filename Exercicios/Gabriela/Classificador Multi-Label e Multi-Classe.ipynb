```python
from sklearn.linear_model import SGDClassifier, RidgeClassifier, LogisticRegression
from sklearn.naive_bayes import GaussianNB, MultinomialNB
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report
from sklearn.multiclass import OneVsRestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix
from sklearn import preprocessing
import matplotlib.pyplot as plt
from scipy.special import expit
from scipy.io import arff
import numpy as np
import pandas as pd
import scipy
```

# Classificador Multi Label 

Único Y, vários rótulos

## MNIST

```python
df = pd.read_csv('./../../mnist_train.csv')
df = df.sample(n=3000)
df
```

```python
df.label.value_counts()
```

```python
X = df.drop(['label'],axis=1)
y = df[['label']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
# Classificadores

# SGD
sgd = SGDClassifier(random_state=42)
# Logistic Regression
lr = LogisticRegression(multi_class='multinomial', solver='lbfgs')
# Naive Bayes
nb = GaussianNB()
# KNN
knn = KNeighborsClassifier(n_neighbors=3)
# Árvore de Decisão
tree = DecisionTreeClassifier()
# Random Forest
rfo = RandomForestClassifier(max_depth=2)
```

#### SGD

```python
sgd.fit(X_train, y_train)
pred_test = sgd.predict(X_test)
cm = confusion_matrix(y_test, pred_test)

print(cm)
print(classification_report(y_test, pred_test))
```

#### Logistic Regression

```python
lr.fit(X_train, y_train)
pred_test = lr.predict(X_test)
cm = confusion_matrix(y_test, pred_test)

print(cm)
print(classification_report(y_test, pred_test)) 
```

#### Naive Bayes

```python
nb.fit(X_train, y_train)
pred_test = nb.predict(X_test)
cm = confusion_matrix(y_test, pred_test)

print(cm)
print(classification_report(y_test, pred_test)) 
```

#### KNN

```python
knn.fit(X_train, y_train)
pred_test = knn.predict(X_test)
cm = confusion_matrix(y_test, pred_test)

print(cm)
print(classification_report(y_test, pred_test)) 
```

#### Árvore de Decisão

```python
tree.fit(X_train, y_train)
pred_test = tree.predict(X_test)
cm = confusion_matrix(y_test, pred_test)

print(cm)
print(classification_report(y_test, pred_test)) 
```

#### Resultados
* Para uma amostragem de 3000 instâncias da base de dados Mnist  
* Melhor método foi o KNN (K=3), apresentando 91% de acurácia e e macro average
    * O dígito 3 apresentou o menor número de acertos, sendo erroneamente classificado cinco vezes como o dígito 2 e quatro vezes como o dígito 5.
    * O classificador teve maior sucesso em identificar o dígito 4, confundindo uma vez com dígito 6 e cinco vezes com o dígito 9

# Classificador Multi Classe

Y com várias colunas 

```python
data, meta = scipy.io.arff.loadarff('./../../scene_arff.arff')
df = pd.DataFrame(data)
df
```

```python
df.columns
```

```python
X = df.drop(['Beach','Sunset','FallFoliage','Field','Mountain','Urban'],axis=1)
Y = df[['Beach','Sunset','FallFoliage','Field','Mountain','Urban']]
```

```python
le = preprocessing.LabelEncoder()
Y['beach'] = le.fit_transform(Y['Beach'])
Y['sunset'] = le.fit_transform(Y['Sunset'])
Y['fallFoliage'] = le.fit_transform(Y['FallFoliage'])
Y['field'] = le.fit_transform(Y['Field'])
Y['mountain'] = le.fit_transform(Y['Mountain'])
Y['urban'] = le.fit_transform(Y['Urban'])

Y = Y.drop(['Beach','Sunset','FallFoliage','Field','Mountain','Urban'], axis=1)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.1)
```

#### Decision Tree 

```python
sgd_clf = DecisionTreeClassifier()

sgd_clf.fit(X_train, Y_train)

Y_pred = sgd_clf.predict(X_test)
print(classification_report(Y_test, Y_pred))
```

### Múltiplas classes por instância

##### Logistic Regression

```python
YY = Y[['sunset']]
```

```python
sgd.fit(X, YY)
pred_test = sgd.predict(X)
print(classification_report(YY, pred_test))
```

##### KNN

```python
knn.fit(X, YY)
pred_test = knn.predict(X)
print(classification_report(YY, pred_test))
```

##### Random Forest

```python
rfo.fit(X, YY)
pred_test = rfo.predict(X)
print(classification_report(YY, pred_test))
```
